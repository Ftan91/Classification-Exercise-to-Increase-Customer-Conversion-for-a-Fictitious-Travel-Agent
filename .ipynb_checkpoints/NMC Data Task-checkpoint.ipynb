## Finding Potential Customers to advertise Carribean Holiday Tour

### Executive Summary
The data used in the task is generated using a quiz similar to [this](http://you.visualdna.com/quiz/whoami#/quiz). The information about the data and methodology is given in the section below. As our client is a Carribean Holiday Operator looking to advertise their trips to potential customers online, it was required to predict the probability of the users in the 'submission.csv' file.

- In selecting the best performing model, all classification scoring metric including Accuracy, Recall, Precision and F1 scores were compared. 'Recall' was prioritized over other metrices because logically speaking, the cost of False Negative outweighs False Positive, representing the loss of potential sale vs loss of marketing cost.
<br><br>

Key Findings:
<br><br>
- Based on the initial clustering exercise, users were classified in 5 clusters. This gives insights into personality characteristics and buying patterns of each distinct groups. Users in cluster 1 tend to avoid purchasing holiday packages which is aligned to several characteristics highlighted in the cluster including users who are introverted in nature and enjoys spending time at home. It is recommended to limiting marketing efforts / budget away from this user cluster.
    
- In the models, it was found that features importance tends to be coherent across the board, primarily having a high budget, being in an older demographic group coupled with high media consumption (ie. tv/concerts/cinema). That said, given the limited marketing budget, it is suggested to target the users for which the predicted probability very high.

### Methodology
1. **Business Understanding**
    - We are consultants hired by as Carribean Holiday Operator to advertise their trips to potential customers online.

<br> <br>

    - We have 4 seperate csv files 'data.csv','taxonomy.csv','conversion.csv','submission.csv'.
    - 'data.csv' contains raw data with user_uids and associated segments, that are identified by an ID corresponding to a label in 'taxonomy.csv. The segments are marked as P (when a user had the opportunity to click on an option in the quiz) and A (when the user actually clicked on an option in the quiz).
    - 'taxonomy.csv' gives category labels for the variables in 'data.csv'.
    - 'conversion.csv' contains a list of customers that have purchased the product in the past.
    - 'submission.csv'contains a list of customers for which predictions have to be made.
    - Calculate and visualize Prevalence to understand the distribution of the options shown to customers and user's response to it.
<br><br>
3. **Data Preparation**  
    - Data Visualisations
    - Data Cleaning
    - Resolving the issue where column marked _A does not have a corresponding _P

    - Identify the clusters in which the users who bought the tour lie.
    - Resolving the problem of class imbalance
<br><br>
4. **Modelling**
    - Checking for possibility of dimentionality reduction using PCA.
    - Building base Model
    - Model Assessment 
    - Removing the columns where user didn't had the opportunity to click on an option in majority of cases (more than 50 percent).
    - Running multiple models and measuring their performance.
    - Hyperparameter tuning
    - Checking for feature importance.
<br><br>
5. **Evaluation**
    - Choosing the best model

#### Transforming the 'score' column in data

```python
#Getting the list of unique columns
all_columns = []
for ele in data['score']:
    d = ast.literal_eval(ele)[0]
    all_columns.extend(list(d.keys()))

all_columns= sorted(list(set(all_columns)))

#Getting the values corresponding to each tag in each row
vals = []
for ele in data['score']:
    d = ast.literal_eval(ele)[0]
    vals.append([ d.get(col, np.nan) for col in all_columns])
    
#Converting to DataFrame
df = pd.DataFrame(data= vals, columns=all_columns)
df.head()
```

### Prevalence

```python
#Converting the tags columns from wide to long format & counting the instances of each tag
prevalence = pd.melt(df).groupby('variable').count().reset_index().rename(
             columns = {'variable':'Label','value':'Count'})
prevalence.head()
```

### Visualizing Prevalence

```python
#Visualizing the percentage prevalence of tags, percentage of times user actually clicked the option when shown. 
prevalence_c = prevalence
prevalence_c['Label'] = prevalence_c['Label'].str.replace('_A','')
prevalence_c['Label'] = prevalence_c['Label'].str.replace('_P','')
prevalence_c['Prevalence'] = np.nan
for i in range(len(prevalence)-1):
    if prevalence_c.loc[i,'Label']==prevalence_c.loc[i+1,'Label']:
        prevalence_c.loc[i,'Prevalence'] = round((prevalence_c.loc[i,'Count']/prevalence_c.loc[i+1,'Count'])*100,2)
prevalence_c.dropna(how='any',inplace=True)
plt.figure(figsize=(8,6))
plt.hist(prevalence_c['Prevalence'],bins=40)
plt.xlabel('Prevalence Percentage')
plt.ylabel('Count')
plt.title('Prevalence of the Tags in percentage')
plt.show()
```

#### Recording the columns marked _A which does not have a corresponding _P 

```python
#List of users for which there is no error in recording the data
customers_with_P = list(prevalence_c['Label']+"_A") #This will be used later

prevalence_c['Label'] = prevalence_c['Label'].astype('int64')
print('There are {} tags that do not have _P, meaning that there is error in recording the data'.format(len(taxonomy)-len(customers_with_P)))
```

## 3. Data Preparation

```python
data = pd.concat([data,df],axis=1).drop('score',axis=1)
data.head()
```

### Removing multiple quiz scores of the same user (keeping the last score)

```python
print(str(len(data['user_uid'].value_counts()[data['user_uid'].value_counts()>1]))+' - Users took quiz for more than once')

#Dropping the rows
data.drop_duplicates('user_uid',keep='last',inplace=True)
```

### 4 possibilities of a tag

1. **Option was shown to the user and user actually actually clicked that option - segment marked as _A is assigned as 1.**
2. **Option was shown to the user and user didn't choose that option - segment marked as _A is assigned as -1.**
3. **Option was not shown to the user - segment marked as _A is assigned as 0.**
4. **Tag that have continous value and user reponded to it, it is left as it is.**

```python
#Seperating continous value columns from the discrete
X = data.iloc[:,3:]
continuous = []
for i in X.columns:
    num = len(X[i].unique())
    if num > 3:
        continuous.append(i)
        
#Creating seperate dataframes for continous and discrete variables
X_disc = X[X.columns[~X.columns.isin(continuous)]]
X_cont = X[X.columns[X.columns.isin(continuous)]].fillna(0)
```

```python
#Cleaning X_disc by assigning -1,0 and 1 on the basis of segment markation
temp = X_disc.copy()
temp.fillna(0, inplace=True)
for col in temp.columns:
    name = col.split('_')
    num = name[0]
    code = name[1]
    if code == 'A':
        if num+'_P' in temp.columns :
            temp[col] = temp[col] + temp[num+'_P']
            temp[col] = temp[col].apply(lambda x: 1 if x==2.0 else (-1 if x==1 else 0))
X_disc=temp
```

### Resolving the issue where column marked _A does not have a corresponding _P

```python
#Joining cleaned data to data & removing the _A columns that don't have a corresponding _P
cleaned_data = data.iloc[:,0:3].join(X_disc.join(X_cont)[customers_with_P])

#Dropping the _P columns from Discrete and Continuous columns
X_disc = remove_p(X_disc)
X_cont = remove_p(X_cont)
```

### Dealing with multicoliniarity

```python
# Using pearson correlation
correlation_matrix = cleaned_data.iloc[:,3:].corr()

#Setting the threshold and removing the columns with high multicoliniarity
correlation_threshold = 0.75
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k = 1).astype(np.bool))
highly_correlated = [column for column in upper.columns if any(upper[column] > correlation_threshold)]
cleaned_data.drop(highly_correlated,axis=1,inplace=True)
print('Dropped {} Columns'.format(len(highly_correlated)))
```

### Clustering the customers

```python
X = cleaned_data.iloc[:,3:]
# Use the ward() function
linkage_array = ward(X)

# Now we plot the dendrogram for the linkage_array containing the distances
# between clusters
label = dendrogram(linkage_array,color_threshold=170)

ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [170, 170], '--', c='k')
plt.xlabel('Data index')
plt.ylabel('Cluster distance')
plt.show()
```

```python
#Retreiving the cluster labels
max_d = 170
cluster_list = fcluster(linkage_array, max_d, criterion='distance')
clusters = len(set(cluster_list))
#Adding the cluster labels to the data
cleaned_data['Cluster'] = cluster_list

#Plotting the clusters
plt.figure(figsize=(8,6))
sns.countplot(cleaned_data['Cluster'])
plt.title('Users per cluster')
plt.ylabel('Number of users')
plt.show()
```

### Discrete tags in  each cluster

```python
pbar=ProgressBar()
X_disc['Labels'] = cleaned_data['Cluster']
cluster_df = pd.DataFrame(index=X_disc.columns[:-1],columns=(range(1,clusters+1)))
for cluster in pbar(range(1,clusters+1)):
    for column in X_disc.columns[:-1]:
        try:
            cluster_df.loc[column,cluster] = X_disc[X_disc['Labels']==cluster][column].value_counts(normalize=True).loc[1]
        except KeyError:
            continue
            
#Renaming the columns
for col in cluster_df.columns:
    cluster_df.rename(columns={col:'Cluster-{}'.format(col)},inplace=True)
#Assigning actual tag as the index by removing _A
cluster_df.index = cluster_df.index.str.replace('_A','').astype('int64')
cluster_df
```

### Identifying top tags for each of the clusters

```python
# #Setting the 'Tag' column in taxonomy as index
taxonomy.set_index('Tag',inplace=True)
taxonomy.index = taxonomy.index.astype('int64')


#Plotting the top tags for each of the clusters
def top_tags(num_of_top,cluster_num):
    top = cluster_df[cluster_num].sort_values(ascending=False)[:num_of_top]
    top_df = taxonomy.loc[top.index].join(top)
    label = top_df.Label.str.split('§')
    labels_to_use = []
    for i in label:
        labels_to_use.append(i[-2]+'--'+i[-1])
    top_df['Label-Last Two'] = labels_to_use
    return make_radar_chart(cluster_num, top_df[cluster_num].values, labels_to_use), display(top_df)

for i in range(1,clusters+1):
    top_tags(10,'Cluster-'+str(i))
```

### Clustering the users who bought the holiday package

```python
plt.figure(figsize=(8,6))
sns.countplot(cleaned_data[cleaned_data['user_uid'].isin(list(conv_users['user_uid']))]['Cluster'])
plt.title('Converted users per cluster')
plt.ylabel('Number of users')
plt.show()
```

> **As we see from the clustering, users who bought the package don't lie in Cluster-1**

### Train Validation Test Split

```python
cleaned_data.drop('Cluster',axis=1,inplace=True)
cleaned_data['Target'] = 0

#Test Data
X_test = cleaned_data[cleaned_data['user_uid'].isin(list(submission['user_uid']))].iloc[:,3:]
X_test.drop('Target',axis=1,inplace=True)
submission_file = cleaned_data[cleaned_data['user_uid'].isin(list(submission['user_uid']))][['user_uid']]
#Remaining Data
X_train = cleaned_data[~cleaned_data['user_uid'].isin(list(submission['user_uid']))]
X_train.loc[X_train['user_uid'].isin(list(conv_users['user_uid'])),'Target']=1
X_train = X_train.iloc[:,3:]
y_train = X_train['Target']
X_train.drop('Target',axis=1,inplace=True)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=1000, random_state=42)
```

### Problem of class imbalance in training data

```python
sns.countplot(y_train)
plt.title('Target Classes in Training Set')
plt.show()
```

#### SMOTE to balance the target variables

```python
sm = SMOTE(random_state = 42, sampling_strategy = 'minority')
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

#Class distribution after SMOTE
sns.countplot(y_train_sm)
plt.title('Target Classes in Training Set')
plt.show()
```

> **X_train with Balanced Class**

### PCA

```python
pca = PCA()

#Scaling the variables
sc = StandardScaler()
X = sc.fit_transform(cleaned_data.iloc[:,3:])
pca.fit_transform(X)

#Plotting the eigen vectors and variance explained
plt.plot(pca.explained_variance_ratio_)
plt.title("Variance Explained, Given Number of Components")
plt.xlabel("Number of Eigenvectors")
plt.ylabel("Variance Explained")
plt.show()
vectors_used = 100
print("If we use",vectors_used,"vectors we can explain",round(pca.explained_variance_ratio_[:vectors_used].sum()*100,2),"% of our data!")
```

## 4. Modelling

### Base Model

```python
logreg = LogisticRegression()
logreg.fit(X_train_sm, y_train_sm)
predictions = logreg.predict(X_val)
print ('Accuracy:', round(accuracy_score(y_val, predictions), 2))
print ('Recall:', round(recall_score(y_val, predictions), 2))
print ('Precision:', round(precision_score(y_val, predictions), 2))
print ('f1:', round(f1_score(y_val, predictions), 2))
```

> **As we are getting such high accuracy, recall and precision. We need to investigate the variables used in the model.**

```python
#Appying Lasso Regularisation on logistic regression to identify the variables being used
C = [0.03, 0.04, 0.05]
for c in C:
    logreg_C = LogisticRegression(penalty = 'l1', C = c, n_jobs = -1, solver = 'saga')
    logreg_C.fit(X_train_sm, y_train_sm)
    predictions = logreg_C.predict(X_val)
    print ('C:', c)
    print ('Number of coef used:', sum(logreg_C.coef_[0] != 0))
    print ('Accuracy:', round(accuracy_score(y_val, predictions), 2))
    print ('Recall:', round(recall_score(y_val, predictions), 2))
    print ('Precision:', round(precision_score(y_val, predictions), 2))
    print ('f1:', round(f1_score(y_val, predictions), 2))
    print ('\n')
```

### Identifying the variables being used in logistic regression with Lasso Regularisation

```python
logreg_base = LogisticRegression(penalty = 'l1', C = 0.03, n_jobs = -1, solver = 'saga')
logreg_base.fit(X_train_sm, y_train_sm)
logreg_coef = X_train.columns[logreg_base.coef_[0] <=0.1]

for i,var in enumerate(logreg_coef):
    try:
        zero_perct = round(X_train[var].value_counts(normalize=True).loc[0]*100,2)
    except KeyError:
        continue
        
print('Variable-',i+1,'has',zero_perct,'% cases where option was not shown')
```

> **As we can see the model is using the variables where the option was not shown to the user. Whereas, our model should actually be based on whether user chose the option or not when he/she was shown that option.**

### Removing the tags where option was not shown in more than 50 percent of cases

```python
#Getting columns list where option was not shown in more than 50 percent of cases
zero_columns = []

#Value count on each column and calculate percent of '0' in that column 
# It represents the cases where option was not shown
for feature in X_train.columns:
    summary = X_train[feature].value_counts(normalize = True)
    try:
        if summary.loc[0] >= 0.5:
            zero_columns.append(feature)
    except KeyError:
        continue
        
#Removing the columns
X_train = X_train[X_train_sm.columns[~X_train_sm.columns.isin(zero_columns)]]
X_train_sm = X_train_sm[X_train_sm.columns[~X_train_sm.columns.isin(zero_columns)]]
X_val = X_val[X_train_sm.columns[~X_train_sm.columns.isin(zero_columns)]]
X_test = X_test[X_train_sm.columns[~X_train_sm.columns.isin(zero_columns)]]
final_columns = X_train.columns
```

### Running Multiple Models

### a) Logistic Regression with GridsearchCV (Lasso Penalty)

```python
#Setting objects and parameters
logreg = LogisticRegression()
params = {'C':[0.03, 0.04, 0.05]}
logreg_cv = GridSearchCV(estimator=logreg_C,param_grid=params,cv=10)
logreg_cv.fit(X_train_sm, y_train_sm)

#Printing best parameters
print('Best Penalty: ',logreg_cv.best_params_)

#Printing performance metrices
print_classification_summary('Logistic Regression with Lasso Penalty', 'Validation Set', logreg_cv.best_estimator_, y_val, X_val)

#Printing the important coefficients
columns_logreg = X_train_sm.columns[(logreg_cv.best_estimator_.coef_ != 0)[0]]
col = [int(item.split("_A")[0]) for item in columns_logreg]
temp_tax = taxonomy.loc[col]
label = temp_tax['Label'].str.split('§')
last_two = []
for i in label:
    last_two.append(i[-2]+'--'+i[-1])
temp_tax['Last Two'] = last_two
temp_tax['Coefficients'] = logreg_cv.best_estimator_.coef_ [logreg_cv.best_estimator_.coef_ != 0]
temp_tax = temp_tax.sort_values('Coefficients',ascending=False)
plt.barh(y=temp_tax['Label'],width=temp_tax['Coefficients'])
plt.xlabel('Coefficients')
plt.title('Coefficients of Tags Used')
plt.show()
```

### b) Ensemble - Random Forest

```python
#Setting rf as an object of RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_sm, y_train_sm)

#Printing performance metrices
print_classification_summary('Random Forest', 'Validation Set', rf, y_val, X_val)

#Printing feature importance
print_feature_importance(taxonomy,rf,final_columns)
```

### c) Ensemble - Random Forest with Hyperparamter Optimization

```python
#Tuning the hyper parameters for Random Forest using GridsearchCV
params_grid = {'max_depth' : range(2, 10, 1),
                 'min_samples_leaf' : range(50,250,50),
                 'max_features': [50, 75, 100]}

rf_grid = RandomForestClassifier()
grid_rf = GridSearchCV(rf_grid, param_grid = params_grid, n_jobs = -1, verbose = True, cv = 10)

grid_rf.fit(X_train_sm, y_train_sm)

print ('Best Paramters: ',grid_rf.best_params_)

#Printing performance metrices
print_classification_summary('Random Forest', 'Validation Set', grid_rf.best_estimator_, y_val, X_val)

#Printing feature importance
print_feature_importance(taxonomy,grid_rf.best_estimator_,final_columns)
```

### d) Ensemble - XGBoost

```python
xgb = XGBClassifier()
xgb.fit(X_train_sm, y_train_sm)

#Printing performance metrices
print_classification_summary('XGBoost', 'Validation Set', xgb, y_val, X_val)

#Printing feature importance
print_feature_importance(taxonomy,xgb,final_columns)
```

## 5. Model Evaluation

```python
#Calculating metrices for the models and assigning to dataframe
index = ['Logistic Regression with GridsearchCV (Lasso Penalty)',
        'Ensemble - Random Forest', 'Ensemble - Random Forest with Hyperparamter Optimization',
        'Ensemble - XGBoost']
models = [logreg_cv.best_estimator_,rf,grid_rf,xgb]
accuracy=[]
precision=[]
recall=[]
f1=[]
for model in models:
    y_pred = model.predict(X_val)
    accuracy.append(round(accuracy_score(y_val,y_pred),2))
    precision.append(round(precision_score(y_val,y_pred),2))
    recall.append(round(recall_score(y_val,y_pred),2))
    f1.append(round(f1_score(y_val,y_pred),2))
score = pd.DataFrame(index=index,columns=['Accuracy','Precision','Recall','F-1 Score'])
score['Accuracy'] = accuracy
score['Precision'] = precision
score['Recall'] = recall
score['F-1 Score'] = f1
score
```

> **All the models (except Random Forest) have similar performance but Logistic regression(with Lasso Regularization) is computationally cheap and would be using only 9 columns instead of 226. So, Logistic regression with a penalty of 0.03 is used to find the final probabilities for the users.** 

```python
test_pred = logreg_cv.best_estimator_.predict_proba(X_test)
submission_file['p_user'] = test_pred[:,1]

#User_ids with highest probability of conversion
submission_file['p_user'] = round(submission_file['p_user'],2)
submission_file = submission_file.sort_values('p_user',ascending=False)
submission_file.to_csv('Ravi_Submission.csv')
```

## 6. Future Task

### Threshold
Formula for calculating the threshold (By Metz (1978), and Zweig and Campbell (1993)) for a classification matrix is given by:
<img src="images/threshold.png">
Here, CFP is cost of False Positive, CTN is cost of True Negative, CFN is cost of False Negative and CTP is cost of True Positive. TPR is True Positive Rate and FPR is False Positive Rate. So, after plugging in the values, we will choose the threshold that gives us maximum fm.


### Calculating threshold by assuming m=1

```python
#Plotting the rec_plot with threshold as annotation
y_pred = logreg_cv.best_estimator_.predict_proba(X_val)[:,1]
roc_plot(y_val,y_pred)

#Getting values of FPR,TPR and Threshold at each point of curve
fpr, tpr, threshold = roc_curve(y_val,y_pred)

#Getting the best threshold
threshold_chosen = 0
difference = 0
for i in range(len(threshold)):
    temp = tpr[i]-fpr[i]
    if temp>difference:
        difference=temp
        threshold_chosen=threshold[i]
threshold_chosen = round(threshold_chosen,2)
print('Best Threshold:- ',threshold_chosen)
```

### Performance metrices

```python
y_pred = np.where(logreg_cv.best_estimator_.predict_proba(X_val)[:,1] > threshold_chosen, 1, 0)
print ('Accuracy score:', round(accuracy_score(y_val, y_pred),2))
print ('Recall score:', round(recall_score(y_val, y_pred), 2))
print ('Precision score:', round(precision_score(y_val, y_pred), 2))
print ('F1 score:', round(f1_score(y_val, y_pred), 2))
```

> In future, m can be calculated using the formula mentioned above. After that, the threshold can be used to evaluate model performance.
